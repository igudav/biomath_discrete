```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from tqdm import tqdm

%matplotlib inline
mpl.rc('text', usetex=True)
mpl.rc('font', size=14)
```

```python
u, r = sp.symbols('u r', nonnegative=True)
f = sp.sqrt(u) * sp.exp(r * (1 - u**2))
f
```

```python
df = f.diff(u)
df_lam = sp.lambdify((u, r), df, 'numpy')
df
```

```python
print(sp.latex(f.diff(u)))
```

```python
f.diff(u).subs(u, 1)
```

```python
t = np.linspace(0, 2, 100)
f1_lam = sp.lambdify((u, r), f, 'numpy')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(9.44882, 3.1))

ax[0].plot(t, f1_lam(t, 0.1))
ax[0].plot([0, 1.2], [0, 1.2], '--')
ax[0].set_xlabel(r'$u$')
ax[0].set_ylabel(r'$f(u)$')
ax[0].annotate(r'$r=0{,}1$', (1, 0.25))

ax[1].plot(t, f1_lam(t, 0.45))
ax[1].plot([0, 1.2], [0, 1.2], '--')
ax[1].set_xlabel(r'$u$')
ax[1].set_ylabel(r'$f(u)$')
ax[1].annotate(r'$r=0{,}45$', (1, 0.25))

ax[2].plot(t, f1_lam(t, 1))
ax[2].plot([0, 1.2], [0, 1.2], '--')
ax[2].set_xlabel(r'$u$')
ax[2].set_ylabel(r'$f(u)$')
ax[2].annotate(r'$r=1$', (1, 0.25))

fig.tight_layout()
```

```python
fig.savefig('../figures/lameray1.svg', bbox_inches='tight')
```

```python
f2_lam = sp.lambdify((u, r), f.subs(u, f), 'numpy')
f3_lam = sp.lambdify((u, r), f.subs(u, f).subs(u, f), 'numpy')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(9.44882, 3.1))
u = np.linspace(0, 3, 100)

u_max = 1.1
u = np.linspace(0, u_max, 100)
ax[0].plot(u, f1_lam(u, 0.1), label=r'$f$')
ax[0].plot(u, f2_lam(u, 0.1), label=r'$f \circ f$')
ax[0].plot(u, f3_lam(u, 0.1), label=r'$f \circ f \circ f$')
ax[0].plot([0, u_max], [0, u_max], '--')
ax[0].set_xlabel(r'$u$')
ax[0].set_ylabel(r'$f(u)$')
ax[0].set_title(r'$r=0{,}1$')

u_max = 1.5
u = np.linspace(0, u_max, 100)
ax[1].plot(u, f1_lam(u, 1))
ax[1].plot(u, f2_lam(u, 1))
ax[1].plot(u, f3_lam(u, 1))
ax[1].plot([0, u_max], [0, u_max], '--')
ax[1].set_xlabel(r'$u$')
ax[1].set_ylabel(r'$f(u)$')
ax[1].set_title(r'$r=1$')

u_max = 2
u = np.linspace(0, u_max, 100)
ax[2].plot(u, f1_lam(u, 1.5))
ax[2].plot(u, f2_lam(u, 1.5))
ax[2].plot(u, f3_lam(u, 1.5))
ax[2].plot([0, u_max], [0, u_max], '--')
ax[2].set_xlabel(r'$u$')
ax[2].set_ylabel(r'$f(u)$')
ax[2].set_title(r'$r=1{,}5$')

fig.tight_layout()
fig.legend(loc='center', bbox_to_anchor=(0.5, 1), ncol=3)
```

```python
fig.savefig('../figures/cycles.pdf', bbox_inches='tight')
```

```python
def bifurc(f, u0, n_converge, n_plot, r_grid):
    res = np.empty((r_grid.shape[0], n_plot))
    for i, r in tqdm(enumerate(r_grid)):
        u = u0
        for j in range(n_converge):
            u = f(u, r)
        for j in range(n_plot):
            res[i][j] = u
            u = f(u, r)
    return res
```

```python
r_num = 500
n_plot = 100
r_grid = np.linspace(0, 3.5, r_num)
R = np.tile(r_grid.reshape(-1, 1), (1, n_plot))
U = bifurc(f1_lam, 0.5, 1000, n_plot, r_grid)
```

```python
fig, ax = plt.subplots(figsize=(9.44882, 4.22441))
ax.plot(R.flatten(), U.flatten(), '.', markersize=0.1)
ax.set_xlabel(r'$r$')
ax.set_ylabel(r'$u$')
```

```python
fig.savefig('../figures/bifurc1.pdf', bbox_inches='tight')
```

```python
def lyap(f, df, u0, n, r_grid):
    M = np.empty((r_grid.shape[0], n))
    M[:, 0] = u0
    for i in tqdm(range(r_grid.shape[0])):
        for j in range(1, n):
            M[i, j] = f(M[i, j-1], r_grid[i])
        M[i] = np.log(np.abs(df(M[i], r_grid[i])))
    return np.mean(M, axis=1)
```

```python
r_num = 1000
n = 1000
r_grid = np.linspace(0, 3.5, r_num)
l = lyap(f1_lam, df_lam, 0.5, n, r_grid)
```

```python
fig, ax = plt.subplots(figsize=(9.44882, 4.22441))
ax.plot(r_grid, l, linewidth=0.7)
ax.plot([0, 3.5], [0, 0], linewidth=0.7)
ax.set_xlim(0, 3.5)
ax.set_xlabel(r'$r$')
ax.set_ylabel(r'$h(u)$')
```
